### Explore Merged df

```python
orders = orders_df[['order_id', 'customer_id', 'order_purchase_timestamp']]
order_items = order_items_df[['order_id', 'product_id', 'order_item_id', 'price']]
order_payments = order_payments_df[['order_id', 'payment_type', 'payment_value']]
customers = customers_df[['customer_id', 'customer_state']]
products = products_df[['product_id', 'product_category_name']]
product_category = product_category_df[['product_category_name', 'product_category_name_english']]
```

```python
# Merge tabel-tabel yang dibutuhkan
merged_df = pd.merge(orders, order_items, on='order_id')
merged_df = pd.merge(merged_df, order_payments, on='order_id')
merged_df = pd.merge(merged_df, customers, on='customer_id')
merged_df = pd.merge(merged_df, products, on='product_id')
merged_df = pd.merge(merged_df, product_category, on='product_category_name')

# Tambahkan kolom 'year'
merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'])
merged_df['year'] = merged_df['order_purchase_timestamp'].dt.year
```

```python
merged_df.info()
```

```python
merged_df.describe()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Kategori Product yang paling banyak terjual

```python
order_items_df.sample(20)
```

```python
product_sales = merged_df.groupby(by='product_category_name_english').agg({"order_item_id": "sum"}).reset_index()
```

```python
product_sales.columns = ['product_category_name_english', 'total_sales']
```

```python
product_sales = product_sales.sort_values(by="total_sales", ascending=False)
```

```python
product_sales.head()
```

```python
top_categories = product_sales.head(15)
```

```python
plt.figure(figsize=(12, 6))
sn.barplot(
    x='total_sales',
    y='product_category_name_english',
    data=top_categories,
    palette='viridis'
)
plt.title('Kategori Produk Paling Banyak Terjual', fontsize=16)
plt.xlabel('Total Penjualan', fontsize=12)
plt.ylabel('Kategori Produk', fontsize=12)
plt.tight_layout()
plt.show()
```

### Pertanyaan 2: Jumlah Pendapatan Per Tahun

```python
merged_df.head()
```

```python
print(merged_df['order_purchase_timestamp'].min())
print(merged_df['order_purchase_timestamp'].max())
```

```python
merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'])
```

```python
merged_df['year'] = merged_df['order_purchase_timestamp'].dt.year
```

```python
revenue_per_year = merged_df.groupby('year')['price'].sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
sn.barplot(x='year', y='price', data=revenue_per_year, palette='viridis')
plt.title('Total Pendapatan per Tahun', fontsize=16)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Total Pendapatan (Skala Logaritmik)', fontsize=12)
plt.yscale('log')  # Mengubah skala sumbu y menjadi logaritmik
plt.tight_layout()
plt.show()
```

### Pertanyaan 3: Distribusi Customer Per State

### Pertanyaan 4: Tipe pembayaran apa yang paling banyak digunakan?

```python
# Menghitung frekuensi setiap tipe pembayaran
payment_type_counts = merged_df['payment_type'].value_counts().reset_index()
payment_type_counts.columns = ['payment_type', 'count']
payment_type_counts.head()
```

```python
plt.figure(figsize=(10, 6))
sn.barplot(x='payment_type', y='count', data=payment_type_counts, palette='viridis')
plt.title('Tipe Pembayaran yang Paling Banyak Digunakan', fontsize=16)
plt.xlabel('Tipe Pembayaran', fontsize=12)
plt.ylabel('Jumlah', fontsize=12)
plt.tight_layout()
plt.show()
```

**Insight:**
- Kategori produk terlaris adalah bed bath table.
- Tahun yang paling banyak mendapatkan pendapatan adalah tahun 2018.
- Sebagian besar customer berasal dari state SP, diikuti oleh RJ dan MG. State-state tersebut merupakan potensi pasar yang besar dan perlu menjadi fokus strategi pemasaran.
- Credit card merupakan metode pembayaran yang paling populer di platform, diikuti oleh boleto dan voucher. Hal ini menunjukkan preferensi customer untuk pembayaran non-tunai

## Conclusion

- Conclution pertanyaan 1 : Berdasarkan analisis, kategori 'bed_bath_table' merupakan kategori produk yang paling banyak terjual.
    Hal ini menunjukkan bahwa produk-produk rumah tangga, khususnya perlengkapan kamar tidur dan kamar mandi,
    sangat diminati oleh konsumen di platform e-commerce ini. Kategori lain yang juga populer adalah
    perlengkapan rumah tangga dan dekorasi.
    
- Conclution pertanyaan 2 : Pendapatan tertinggi terjadi pada tahun 2018, yang menunjukkan tren pertumbuhan positif dari tahun-tahun
    sebelumnya. Hal ini mengindikasikan bahwa platform e-commerce ini mengalami perkembangan bisnis yang baik
    dan semakin banyak konsumen yang melakukan pembelian.

- Conclution pertanyaan 3 : Sebagian besar pelanggan berasal dari negara bagian SP (São Paulo), diikuti oleh RJ (Rio de Janeiro)
    dan MG (Minas Gerais). Hal ini sesuai dengan distribusi populasi di Brasil, di mana São Paulo merupakan
    negara bagian dengan populasi terbesar. Konsentrasi pelanggan di tiga negara bagian ini juga
    menunjukkan potensi ekspansi pasar ke negara bagian lain yang masih memiliki penetrasi rendah.

- Conclution pertanyaan 4 : Kartu kredit (credit_card) merupakan metode pembayaran yang paling banyak digunakan oleh pelanggan,
    diikuti oleh boleto (sistem pembayaran khas Brasil) dan voucher. Hal ini mencerminkan preferensi
    konsumen Brasil untuk metode pembayaran non-tunai dan mengindikasikan tingkat adopsi teknologi
    finansial yang cukup tinggi di antara pengguna platform.
